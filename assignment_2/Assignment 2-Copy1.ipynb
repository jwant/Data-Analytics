```python
mean = df['Age'].mean()
std = df['Age'].std()
df['age_z_score'] = df.apply(lambda a: (a['Age'] - mean)/std, axis=1)
```

```python
def return_age_group(age):
    if age <= 12:
        return 'Child'
    elif 13 <= age <= 19:
        return 'Teenager'
    elif 20 <= age <= 39:
        return 'Young Adult'
    elif 40 <= age <= 59:
        return 'Middle Aged'
    else:
        return 'Old Aged'
    
df['age_group'] = df.apply(lambda a: return_age_group(a['Age']), axis=1)


plot = px.pie(df,'age_group')
plot.show()
```
